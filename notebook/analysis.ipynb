```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Cell 2: Load Dataset
df = pd.read_csv("../data/creditcard.csv")
df.head()
```

```python
# Cell 3: Basic Info
print("Shape:", df.shape)
df.info()
```

```python
# Cell 4: Missing Values
df.isnull().sum()
```

```python
# Cell 5: Class Distribution
sns.countplot(x='Class', data=df)
plt.title("Class Distribution (0 = Legit, 1 = Fraud)")
plt.show()

print(df['Class'].value_counts(normalize=True))
```

```python
# Cell 6: Summary Statistics
df.describe()
```

```python
# Cell 7: Time and Amount Distribution
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
sns.histplot(df['Time'], bins=100, ax=axes[0], kde=True)
axes[0].set_title("Distribution of Time")
sns.histplot(df['Amount'], bins=100, ax=axes[1], kde=True)
axes[1].set_title("Distribution of Amount")
plt.tight_layout()
plt.show()
```

```python
# Cell 8: Correlation Matrix
plt.figure(figsize=(12,10))
corr = df.corr()
sns.heatmap(corr, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Heatmap of All Features")
plt.show()
```

```python
# Cell 9: Correlation with Class
corr_target = df.corr()['Class'].sort_values(ascending=False)
print("Features most correlated with fraud:\n")
print(corr_target.head(10))
```

```python
# Cell 10: PCA Visualization (optional if you want dimensionality reduction)
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

scaled_df = df.copy()
scaled_df[['Time', 'Amount']] = StandardScaler().fit_transform(scaled_df[['Time', 'Amount']])
X = scaled_df.drop(columns=['Class'])
y = df['Class']

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

plt.figure(figsize=(10,6))
plt.scatter(X_pca[y == 0][:, 0], X_pca[y == 0][:, 1], label='Legit', alpha=0.5, s=2)
plt.scatter(X_pca[y == 1][:, 0], X_pca[y == 1][:, 1], label='Fraud', alpha=0.8, s=10, c='red')
plt.legend()
plt.title("PCA Projection of Transactions")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.show()
```
